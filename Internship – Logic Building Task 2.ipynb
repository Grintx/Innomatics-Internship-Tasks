```python
# Problem 1: Unique Words in a Sentence

sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Output
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print(f"Highest Salary: {highest_employee} - {employees[highest_employee]}")
```

```python
# Problem 3: Maximum and Minimum

numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Problem 4: Count Products Above 1000

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

```python
# Problem 5: Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Unique phone numbers:", unique_numbers)
```

```python
# Problem 7: Character Frequency

text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print(frequency)
```

```python
# Problem 8: Convert List to Tuple

data_list = [10, 20, 30]

data_tuple = tuple(data_list)

print("Tuple:", data_tuple)
```

```python
# Problem 9: Check Key Existence

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

key_to_check = "Ravi"

if key_to_check in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

```python
# Problem 10: Calculate Average Marks

marks = [70, 75, 80, 65]

average = sum(marks) / len(marks)

print("Average Marks:", average)
```

```python
# Problem 1: Smart Parking Lot Management System

def manage_parking(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0

    # Process vehicle logs
    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)

    if current_vehicles > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


# Example usage
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]
manage_parking(capacity, logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # 5 minutes per km

    # Traffic impact
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 8

    # Weather impact
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Example usage
estimate_delivery_time(8, "High", "Rainy")
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)

    occupancy_percentage = (booked_count / total_seats) * 100
    print("Occupancy:", round(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    if occupancy_percentage >= 90:
        print("Suggestion: Consider Opening Additional Show")


# Example usage
total = 200
booked = [1] * 150  # 150 seats booked
analyze_occupancy(total, booked)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    print("Average CPU Load:", round(average_cpu), "%")

    if average_cpu < 50:
        print("Server Status: Normal")
    elif average_cpu <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Example usage
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor

def monitor_resource_usage(resource_usage):
    overused = []

    # Threshold for overuse (example: more than 8 hours)
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Example usage
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resource_usage(resources)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def manage_event_registration(capacity, registrations):
    
    confirmed = 0
    waitlisted = 0

    # Process registrations using loop
    for _ in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Example usage
manage_event_registration(100, 105)
```

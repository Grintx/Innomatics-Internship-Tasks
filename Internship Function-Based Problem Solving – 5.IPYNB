```python
# Problem 1: Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    total_likes = 0
    
    # Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes = [200, 150, 300, 400]
analyze_post_engagement(likes)
```

```python
# Problem 3: Rainfall Adequacy Checker

def rainfall_analysis(rainfall_data, required_level):
    total = 0

    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    print("Average Rainfall:", average)

    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
rainfall = [70, 75, 80, 65]
rainfall_analysis(rainfall, 70)
```

```python
# Problem 4: Duplicate Account Detection

def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
users = ["user1", "user2", "user1", "user3"]
detect_duplicate_accounts(users)
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(prices):
    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: Application Health Checker

def check_system_health(error_count):
    print("Error Count:", error_count)

    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example usage
check_system_health(7)
```

```python
# Problem 9: Attendance Eligibility System

def check_exam_eligibility(attendance_list):
    present_days = 0

    for day in attendance_list:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example usage
attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    # Slab 1: First 100 units
    if units <= 100:
        bill = units * 3

    # Slab 2: 101â€“200 units
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)

    # Slab 3: Above 200 units
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example usage
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
